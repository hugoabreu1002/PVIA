```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (12, 6), dpi = 300)
axis = fig.add_subplot(1,1,1, title = 'Gerao em Mdia Mensal (MW)')

y = np.random.randn(10)
t = pd.date_range(start='2018-03-27', periods=10, freq='D')
t = list(map(lambda x: x[5:], list(t.astype('str'))))

axis.plot(t, y)
#axis.set_xticks(t)
axis.tick_params(axis='x', labelrotation=90)

plt.show()
```
